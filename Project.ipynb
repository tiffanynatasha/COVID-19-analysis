```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid_cases = pd.read_csv("Covid-19 Cases.csv")
covid_cases
```

```python
covid_cases['year'] = pd.DatetimeIndex(covid_cases['date']).year
covid_cases['month'] = pd.DatetimeIndex(covid_cases['date']).month
covid_cases
```

```python
grouped_date = covid_cases.groupby(['month', 'year']).sum()
grouped_date = grouped_date[['cases']]
#covid_cases['month'] = pd.to_datetime(covid_cases['month'], format='%m').dt.month_name()
grouped_date = grouped_date.reset_index()
grouped_date["month"] = grouped_date["month"].astype("int")
grouped_date["year"] = grouped_date["year"].astype("int")
grouped_date.sort_values(by = ["year", "month"])
grouped_date
```

```python
grouped_date['Date'] = pd.to_datetime(grouped_date[['year', 'month']].assign(DAY=1))
grouped_date = grouped_date.set_index("Date")
grouped_date = grouped_date[["cases"]]
grouped_date.plot()
```

```python
grouped_region = covid_cases.groupby("area").sum()
grouped_region = grouped_region[["cases"]]
grouped_region.plot(kind = "bar")
```
